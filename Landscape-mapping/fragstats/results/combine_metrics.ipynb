```python
import pandas as pd

# List of station names
stations = ['station1', 'station2', 'station3', 'station4', 'station5', 'station6', 'station7', 'station8']

# Initialize an empty dataframe to store the combined data for all stations
combined_data_all = pd.DataFrame()

for i, station in enumerate(stations, start=1):
    # Load the class and land data for the current station
    data_class = pd.read_csv(f'{station}_class.csv')
    data_land = pd.read_csv(f'{station}_land.csv')

    # Get the length of the class dataset
    len_data_class = len(data_class)

    # Replicate the row in data_land to match the length of data_class
    data_land_repeated = pd.concat([data_land]*len_data_class, ignore_index=True)

    # Join the two dataframes for the current station
    combined_data = pd.concat([data_class, data_land_repeated], axis=1)

    # Change the first column name to 'Station' and set its values to the current station number
    combined_data.rename(columns={combined_data.columns[0]: 'Station'}, inplace=True)
    combined_data['Station'] = i

    # Append the data for the current station to the combined data for all stations
    combined_data_all = combined_data_all.append(combined_data, ignore_index=True)
# Fill NaN values with 0
combined_data_all.fillna(0, inplace=True)

# Save the combined dataframe to a CSV file
combined_data_all.to_csv('combined_data.csv', index=False)

combined_data_all.tail()

```
